```python
import pandas as pd
import matplotlib.pyplot as plt
```

### READING DATASET

```python
df = pd.read_excel("./State Occupational Deaths by Event.xlsx")

df
```

### CHANGING COLUMN NAMES TO LOWER CASE

```python
df= df.rename(columns=str.lower)

df
```

### Filling Null values with 0

```python
df.fillna(0,inplace=True)

df
```

### Renaming Columns

```python
df.rename(columns={"deaths" : "fatality",
                   "rate (e)" : "fatality_rate",
                   "transportation incidents (c)": "Transportation",
                   "violence and other injuries by persons or animals (d)" : "Violence",
                   "falls, slips, or trips" : "Fall/Slip",
                   "contact with objects and equipment" : "Objects/Equipments",
                   "exposure to harmful substances or environments" : "Exposure",
                   "fires and explosions" : "Fire/Explosion"},inplace=True)

df
```

### Sorting Values

```python
df.sort_values(["state","year"], ascending=[True, False],inplace=True)

df
```

### Formulating new column called Unknown

```python
df["Unknown"] = df["fatality"] - (df["Transportation"] + df["Violence"] + df["Fall/Slip"] + df["Objects/Equipments"] + df["Exposure"] + df["Fire/Explosion"])
```

```python
df
```

```python
#converting data types of columns to int
df = df.astype({"Transportation" : "int","Violence" : "int", "Fall/Slip" : "int","Objects/Equipments" : "int", "Exposure" : "int", "Fire/Explosion" : "int", "Unknown" : "int"})

df.info()
```

### Melting the Dataframe

```python
 df_melted = df.melt(id_vars=["state","year"],
         value_vars=df.iloc[:, 4:],
         var_name="types of events",
         value_name="fatality_count")
```

```python
df_melted.head(20)
```

```python
# saving locally
df_melted.to_csv("./fatality_cleaned.csv",index=False)
```
